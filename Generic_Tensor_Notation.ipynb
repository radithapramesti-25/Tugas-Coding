## Generic Tensor Notation

Sebagai contoh, tensor rank 4 umum digunakan untuk gambar, di mana setiap dimensinya sesuai dengan:

1. Jumlah gambar dalam batch pelatihan, misalnya, 32
2. Tinggi gambar dalam piksel, misalnya, 28 untuk angka MNIST.
3. Lebar gambar dalam piksel, misalnya, 28
4. Jumlah saluran warna, misalnya, 3 untuk gambar berwarna penuh (RGB)

```python
import torch
import tensorflow as tf
```

```python
# gambar dalam pytorch
images_pt = torch.zeros([32, 28, 28, 3])
```

```python
# gambar dalam pixel
images_pt
```

```python
# gambar dalam tensor flow
images_tf = tf.zeros([32, 28, 28, 3])
```

```python
# gambar dalam pixel
images_tf
```
