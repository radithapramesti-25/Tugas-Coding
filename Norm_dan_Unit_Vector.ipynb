## Unit Vector

```python
# membuat vector v dengan elemen (3,4)
v = np.array([3,4])
```

```python
# menampilkan unit vector dari vector v
v
```

```python
# menghitung nilai L2 norm dari vector v
(3**2 + 4**2)**(1/2)
```

```python
# menghitung panjang unit
norm = np.linalg.norm(v)
norm
```
