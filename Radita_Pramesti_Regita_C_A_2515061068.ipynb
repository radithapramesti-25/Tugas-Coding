```python
# melakukan import tensorflow
import tensorflow as tf
```

```python
# menampilkan variable x tensor flow
x_tf = tf.Variable(30, dtype=tf.int16)
x_tf
```

```python
# menampilkan tipe data variable x tensorflow
x_tf
```

```python
# menampilkan tensorshape
x_tf.shape
```

```python
# menampilkan variable y tensorflow
y_tf = tf.Variable(3, dtype=tf.int16)
```

```python
# menjumlahkan variable x dan y tensorflow
x_tf + y_tf
```

```python
# cara lain menjumlah variable x dan y tensorflow
tf_sum = tf.add(x_tf, y_tf)
tf_sum
```

```python
# jumlah numpy tensorflow
tf_sum.numpy()
```

```python
# type dari jumlah tensorflow
type(tf_sum.numpy())
```
