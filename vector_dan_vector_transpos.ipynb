## Vectors

```python
import numpy as np
```

```python
# vector x dengan ukuran 3x1
x = np.array([10, 2, 4])
```

```python
# menghitung panjang vector x
len(x)
```

```python
# mengetahui ukuran vector x
x.shape
```

```python
# menampilkan tipe data dari vector x
type(x)
```

```python
# mengakses elemen pertama dari vector x
x[0]
```

```python
# menampilkan tipe data elem pertama dari vector x
type(x[0])
```

## Vector Transpos

```python
# melakukan transpose pada vector x dan simpan pada variabel x_t
x_t = x.T
x_t
```

```python
# mengetahui ukuran transpose vector x
x_t.shape
```

```python
# Vector y dengan ukuran 3 x 1
y = np.array([[30, 3, 5]])
y
```

```python
# mengetahui ukuran vector y
y.shape
```

```python
# melakukan transpose pada vector y dan simpan pada variabel y_t
y_t = y.T
y_t
```

```python
# mengetahui ukuran transpose vector y
y_t.shape
```
