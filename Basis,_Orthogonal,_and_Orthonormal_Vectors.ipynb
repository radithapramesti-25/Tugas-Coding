```python
import numpy as np
```

```python
# membuat vector i dengan elemen (1,0)
i = np.array([1, 0])
i
```

```python
# membuat vector y dengan elemen (0, 1)
j = np.array([0, 1])
j
```

```python
# membuktikan bahwa vector i dan j orthogonal
np.dot(i, j)
```

```python
# membuat vector v dengan elemen (1,2)
v = np.array([1, 2])
v
```

```python
# membuat vector w dengan elemen (2,-1)
w = np.array([2, -1])
w
```

```python
# membuktikan bahwa vector v dan w orthogonal
np.dot(v, w)
```
